# Introduction

## MAGIC – Major Atmospheric Gamma Imaging Cherenkov Telescope

## MAGIC (Major Atmospheric Gamma Imaging Cherenkov)

MAGIC is a system consisting of two Cherenkov telescopes, each with a diameter of 17 meters. They are designed to observe gamma rays from galactic and extragalactic sources in the very high energy range (from 30 GeV to 100 TeV).

Currently, the MAGIC telescopes are operated by approximately 165 astrophysicists from 24 institutions and consortia across 12 countries. MAGIC has enabled the discovery and study of new classes of gamma-ray sources, such as pulsars and gamma-ray bursts (GRBs).

## Particles from outer space

Cosmic particles, $\gamma$-quanta (photons) and hadrons (protons), interact with the atmosphere, generating showers of secondary particles. Moving at nearly the speed of light, these particles emit Cherenkov radiation. The telescopes capture images of this radiation. By analyzing these images, it is possible to determine the type of cosmic particle: photon or proton.

<center><img src="https://github.com/hse-ds/iad-applied-ds/blob/master/2021/hw/hw1/img/shower.jpg?raw=1" width="500"></center>









## Photos

The task of the atmospheric Cherenkov telescope is to capture the shower image by measuring the Cherenkov light from the shower particles. This image represents a geometric projection of the shower onto the detector. To analyze these images, image parameters or so-called Hillas parameters were introduced. There are two types of image parameters: shape parameters and orientation parameters. (Source: http://ihp-lx.ethz.ch/Stamet/magic/parameters.html)

# Conditional WGAN

We will use `Conditional WGAN`, which is shown in the figure. As the condition `y`, we will use the class label: **0** - hadron, **1** - photon. Thus, we will inform the generator for which particle it needs to generate the image parameters.

<center><img src="https://github.com/hse-ds/iad-applied-ds/blob/master/2021/hw/hw1/img/cgan.png?raw=1" width="800"></center>


The generator $\hat{x} = G(z, y)$ will take as input a noise vector $z$ and a condition vector $y$, and will output the generated (fake) parameter vector $\hat{x}$.

The discriminator $D(x, y)$ will take as input the parameter vector $x$ and the condition vector $y$, and will output a rational number.

We will train the `Conditional WGAN` with the following loss function:

$$L(G, D) = -\frac{1}{n} \sum_{x_i \in X, y_i \in y} D(x_i, y_i) + -\frac{1}{n} \sum_{z_i \in Z, y_i \in y} D(G(z_i, y_i), y_i) \to \max_G \min_D$$

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
from torch.autograd import Variable

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

## Task 2 (1 point)

Implement a neural network for the generator with the following layers:
- Fully connected layer with 100 neurons;
- Batch normalization layer;
- ReLU activation function;
- Fully connected layer with 100 neurons;
- Batch normalization layer;
- ReLU activation function;
- Output layer.

Hint: Use the `nn.Sequential()` function.

```python
class Generator(nn.Module):
    
    def __init__(self, n_inputs, n_outputs):
        super(Generator, self).__init__()
        
        ### YOUR CODE IS HERE ######
        self.n_outputs = n_outputs
        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, n_outputs)
        )
        ### THE END OF YOUR CODE ###

    def forward(self, z, y):
        zy = torch.cat((z, y), dim=1)
        return self.net(zy).view(-1, self.n_outputs)
```

## Task 3 (1 point)

Implement a neural network for the discriminator with the following layers:
- Fully connected layer with 100 neurons;
- ReLU activation function;
- Fully connected layer with 100 neurons;
- ReLU activation function;
- Output layer.

Hint: Use the `nn.Sequential()` function.

```python
class Discriminator(nn.Module):
    
    def __init__(self, n_inputs):
        super(Discriminator, self).__init__()
        
        ### YOUR CODE IS HERE ######
        self.n_inputs = n_inputs
        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 1)
        )
        ### THE END OF YOUR CODE ###

    def forward(self, x, y):
      for p in self.parameters():
        p.data.clamp_(-0.01, 0.01)
      xy = torch.cat((x, y), dim=1)
      return torch.sigmoid(self.net(xy))
```

## Task 4 (2 points)

Implement a class for training the generative model.

- Hint 1: Don't forget to limit the weights of the discriminator. Use `p.data.clamp_(-0.01, 0.01)`, where `p` represents the discriminator's weights.
- Hint 2: `n_critic` - the number of discriminator training iterations per one generator training iteration.
- Hint 3: Use `X_tensor = torch.tensor(X_numpy, dtype=torch.float, device=DEVICE)` to convert numpy arrays to tensors.

```python
# for i, (real_batch, cond_batch) in enumerate(DataLoader
#                                              (TensorDataset(torch.tensor(X_train, dtype=torch.float, device=DEVICE),
#                                                             torch.tensor(y_train, dtype=torch.float, device=DEVICE)), 8,
#                                                shuffle=True)):
#   print('i:', i) 
#   print('real_batch:', real_batch)
#   print('cond_batch:', cond_batch)
#   break
```

```python
from torch.types import Device
class Fitter(object):
  # функции потерь пока из сема, а не из вышенаписанной формулы, поскольку по ней лосс растет с каждой новой эпохой.
    
    def __init__(self, generator, discriminator, prior, batch_size=32, n_epochs=10, latent_dim=1, lr=0.0001, n_critic=5):
        
        self.generator = generator
        self.discriminator = discriminator
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.n_critic = n_critic
        self.prior = prior
        
        self.opt_gen  = torch.optim.RMSprop(self.generator.parameters(), lr=self.lr)
        self.opt_disc = torch.optim.RMSprop(self.discriminator.parameters(), lr=self.lr)
        
        self.generator.to(DEVICE)
        self.discriminator.to(DEVICE)
    
    
    def fit(self, X, y):
        
        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)
        
        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)
        
        # Turn on training
        self.generator.train(True)
        self.discriminator.train(True)
        self.gen_loss_trace = []
        self.discr_loss_trace = []
        self.loss_history = []

        iter_i = 0

        # Fit GAN
        for epoch in range(self.n_epochs):
          for i, (real_batch, cond_batch) in enumerate(DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)):
                
            ### YOUR CODE IS HERE ######
            real_batch = real_batch.to(DEVICE)
            cond_batch = cond_batch.to(DEVICE)
            
            # генерируем новые объекты
            num_objects = real_batch.shape[0]
            z = self.prior.sample((num_objects, ))
            gen_objects = self.generator(z, cond_batch)

            # считаем прогнозы дискриминатора для реальных и фейковых объектов
            real_objects_scores, gen_objects_scores = torch.split(self.discriminator(
            torch.cat([real_batch, gen_objects], dim=0), 
            torch.cat([cond_batch, cond_batch], dim=0)), num_objects)
            if (iter_i % (1 + self.n_critic)) < 1:
              # делаем шаг обучения генератора
              self.opt_gen.zero_grad()
              gen_loss = -gen_objects_scores.mean()
              gen_loss.backward()
              self.opt_gen.step()
              self.gen_loss_trace.append((iter_i, gen_loss.item()))
            else:
              # делаем шаг обучения дискриминатора
              self.opt_disc.zero_grad()
              discr_loss = gen_objects_scores.mean() - real_objects_scores.mean()
              discr_loss.backward()
              self.opt_disc.step()
              self.discr_loss_trace.append((iter_i, discr_loss.item()))

            iter_i += 1
                
            ### THE END OF YOUR CODE ###
                    
          # caiculate and store loss after an epoch
          Z_noise = torch.normal(0, 1, (len(X_real), self.latent_dim), device=DEVICE)
          X_fake = self.generator(Z_noise, y_cond)
          loss_epoch = torch.mean(self.discriminator(X_real, y_cond)) - torch.mean(self.discriminator(X_fake, y_cond))
          self.loss_history.append(loss_epoch.detach().cpu())
                    
        # Turn off training
        self.generator.train(False)
        self.discriminator.train(False)
```

## Training

```python
%%time
latent_dim = 10
prior = torch.distributions.Normal(torch.zeros(10).cuda(), torch.ones(10).cuda())
generator = Generator(n_inputs=latent_dim+y.shape[1], 
                          n_outputs=X_train.shape[1])
discriminator = Discriminator(n_inputs=X_train.shape[1]+y.shape[1])

fitter = Fitter(generator, discriminator, prior, batch_size=50, n_epochs=100, latent_dim=latent_dim, lr=0.0001, n_critic=5)
fitter.fit(X_train, y_train)
```

```python
# WGAN learning curve
plt.figure(figsize=(9, 5))
plt.plot(fitter.loss_history)
plt.xlabel("Epoch Number", size=14)
plt.ylabel("Loss Function", size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.title("Conditional WGAN Learning Curve", size=14)
plt.grid(b=1, linestyle='--', linewidth=0.5, color='0.5')
plt.show()
```

## Task 6 (1 point)

Implement a neural network for the decoder with the following layers:
- Fully connected layer with 100 neurons;
- Batch normalization layer;
- ReLU activation function;
- Fully connected layer with 100 neurons;
- Batch normalization layer;
- ReLU activation function;
- Output layer for mu; Output layer for log_sigma;

Hint: use the function `nn.Sequential()`.

```python
class Encoder(nn.Module):
    def __init__(self, n_inputs, lat_size):
        super(Encoder, self).__init__()
        
        ### YOUR CODE IS HERE ######
        self.lat_size = lat_size
        self.enc_net = self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU()
        )
        
        self.mu = nn.Linear(100, self.lat_size)
        self.log_sigma = nn.Linear(100, self.lat_size)
        ### THE END OF YOUR CODE ###
        
    def forward(self, x, y):
        z = torch.cat((x, y), dim=1)
        z = self.enc_net(z)
        mu = self.mu(z)
        log_sigma = self.log_sigma(z)
        return mu, log_sigma 
```

## Task 7 (1 point)

Implement a neural network for the decoder with the following layers:
- Fully connected layer with 100 neurons;
- ReLU activation function;
- Fully connected layer with 100 neurons;
- ReLU activation function;
- Output layer.

Hint: use the function `nn.Sequential()`.

```python
class Decoder(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super(Decoder, self).__init__()
        
        ### YOUR CODE IS HERE ######
        self.dec_net = nn.Sequential(
            nn.Linear(n_inputs, 100), 
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, n_outputs)
        )
            # в семинаре используется nn.Sigmoid(), но с ней лосс больше + распределения на картинках ужасные + ROC_AUC = 0.99

        ### THE END OF YOUR CODE ###
        
    def forward(self, z, y):
        z_cond = torch.cat((z, y), dim=1)
        x_rec = self.dec_net(z_cond)
        return x_rec
```

## Task 8 (1 point)

Implement a class for training a variational autoencoder.

Hint: Use `X_tensor = torch.tensor(X_numpy, dtype=torch.float, device=DEVICE)` to convert numpy to tensor.

```python
class VAEFitter(object):
    
    def __init__(self, encoder, decoder, batch_size=32, n_epochs=10, latent_dim=1, lr=0.0001, KL_weight=0.001):
        
        self.encoder = encoder
        self.decoder = decoder
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.KL_weight = KL_weight
        
        self.criterion = nn.MSELoss()
        self.opt  = torch.optim.RMSprop(list(self.encoder.parameters()) + list(self.decoder.parameters()), lr=self.lr)
        
        self.encoder.to(DEVICE)
        self.decoder.to(DEVICE)
        
        
    def sample_z(self, mu, log_sigma):
        eps = torch.randn(mu.shape).to(DEVICE)
        return mu + torch.exp(log_sigma / 2) * eps
    
    def custom_loss(self, x, rec_x, mu, log_sigma):
        KL = torch.mean(-0.5 * torch.sum(1 + log_sigma - mu ** 2 - log_sigma.exp(), dim = 1), dim = 0)
        recon_loss = self.criterion(x, rec_x)
        return KL*self.KL_weight + recon_loss
    
    
    def compute_loss(self, x_batch, cond_batch):
        
        ### YOUR CODE IS HERE ######

        mu, log_sigma = self.encoder(x_batch, cond_batch)
        z = self.sample_z(mu, log_sigma)
        reconstructed_x = self.decoder(z, cond_batch)
        
        loss = self.custom_loss(x_batch, reconstructed_x, mu, log_sigma)
        
        ### THE END OF YOUR CODE ###

        return loss
    
    
    def fit(self, X, y):
        
        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)
        
        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)
        
        # Turn on training
        self.encoder.train(True)
        self.decoder.train(True)
        
        self.loss_history = []

        # Fit VAE
        for epoch in range(self.n_epochs):
            for i, (x_batch, cond_batch) in enumerate(DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)):
                
                x_batch = x_batch.to(DEVICE)
                cond_batch = cond_batch.to(DEVICE)

                # caiculate loss
                loss = self.compute_loss(x_batch, cond_batch)
                
                # optimization step
                self.opt.zero_grad()
                loss.backward()
                self.opt.step()
                    
            # calculate and store loss after an epoch
            loss_epoch = self.compute_loss(X_real, y_cond)
            self.loss_history.append(loss_epoch.detach().cpu())
                    
        # Turn off training
        self.encoder.train(False)
        self.decoder.train(False)
```

## Training
Обучим модель на данных.

```python
%%time

latent_dim = 10

encoder = Encoder(n_inputs=X_train.shape[1]+y.shape[1], lat_size=latent_dim)
decoder = Decoder(n_inputs=latent_dim+y.shape[1], n_outputs=X_train.shape[1])

vae_fitter = VAEFitter(encoder, decoder, batch_size=50, n_epochs=100, latent_dim=latent_dim, lr=0.001, KL_weight=0.001)
vae_fitter.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# обучаем классификатор
clf = GradientBoostingClassifier()
clf.fit(XX_train, yy_train)

# получаем прогнозы
yy_test_proba = clf.predict_proba(XX_test)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(yy_test, yy_test_proba)
print("ROC AUC = ", auc)
```

For the CVAE, we obtained a ROC AUC of around 0.7 (lower is better). Thus, we see that in this task both models behave similarly. But maybe there's a way to improve them? Or is there another model? :)
